```python
from datasets import load_dataset
from os import cpu_count
from json import loads


task_id2test = {}
num_proc = cpu_count()
data = load_dataset("mbpp", split="train", num_proc=num_proc) ########

for datum in data:
    task_id = datum['task_id']
    datum.pop('task_id')
    task_id2test[task_id] = datum

task_id2samples = {}

with open('MBPP(Python)_nucleus92_3attempts.jsonl', 'r') as f:
    for line in f:
        datum = loads(line)
        datum.pop('attempt')
        sample = datum['sample']
        datum.pop('sample')
        task_id = datum['task_id']
        datum.pop('task_id')
        if task_id not in task_id2samples:
            task_id2samples[task_id] = [[datum]]
        elif sample >= len(task_id2samples[task_id]):
            task_id2samples[task_id].append([datum])
        else:
            task_id2samples[task_id][sample].append(datum)

task_id2pair = {}

# for task_id, samples in task_id2samples.items():
#     for sample in samples:
#         if sample[-1]['compilable']:
#             logger.info(f"testing " + samples[-1]['generation'])
#             exec(sample[-1]['generation'])
#             exec(task_id2test[task_id]['test_setup_code'])
#             test_list = task_id2test[task_id]['test_list']
#             for test in test_list:
#                 exec(test)
```

```python
import contextlib
import signal


class TimeoutException(Exception):
    pass


@contextlib.contextmanager
def time_limit(seconds: float):
    def signal_handler(signum, frame):
        raise TimeoutException("Timed out!")

    signal.setitimer(signal.ITIMER_REAL, seconds)
    signal.signal(signal.SIGALRM, signal_handler)
    try:
        yield
    finally:
        signal.setitimer(signal.ITIMER_REAL, 0)


for task_id, samples in task_id2samples.items():
    test_setup_code = task_id2test[task_id]['test_setup_code']
    for sample in samples:
        if sample[-1]['compilable']:
            generation = sample[-1]['generation']
            test_code = generation
            if len(test_setup_code) > 0:
                test_code += "\n\n" + test_setup_code
            test_list = task_id2test[task_id]['test_list']
            try:
                for test in test_list:
                    with time_limit(3.):
                        exec(test_code + "\n\n" + test, {})
                print('passed')
            except AssertionError as e:
                print("failed: AssertionError")
            except BaseException as e:
                print(f"failed: {e}")
        else:
            pass
```
